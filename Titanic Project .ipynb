```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex',]].apply(male_female_child,axis=1)
```

```python
#2) What deck were the passengers on and how does that relate to their class?
```

```python
titanic_df.head()
```

```python
#cabin column is information on the deck
#Cabin column has lots of null value and we'll gonna drop them
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
#we only need character like A.B,C,D,E no numbers
```

```python
levels = []
for level in deck:
    levels.append(level[0])
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
```

```python
cabin_df
sns.countplot('Cabin',data=cabin_df,palette='summer_d')
#if looking up matplotlib plus colormaps, can get a range of palette that can be used. 
#https://matplotlib.org/users/colormaps.html
#remeber to add '_d' to the color name, liike 'summer_d'
```

```python
titanic_df.head()
```

```python
# 3) where did the passengers come from?
sns.countplot('Embarked',data=titanic_df,hue='Pclass')
#change sns.factorplot() sns.countplot()
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
```

```python
titanic_df['Alone']
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] >0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
titanic_df['Alone']
```

```python
sns.countplot("Alone",data=titanic_df,palette='Blues')
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'no',1:'yes'})
```

```python
titanic_df['Survivor']
```

```python
sns.countplot('Survivor', data=titanic_df, palette='Set1')
```

```python
sns.factorplot('Pclass','Survived',hue = 'person',data=titanic_df)
```

```python
sns.lmplot('Age','Survived',data=titanic_df)
```

```python
sns.lmplot('Age','Survived',hue = 'Pclass',data=titanic_df,palette='winter')
```
